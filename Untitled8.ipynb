```python
class ItemDownload():
    '''
    keep tracking of downloading file and manage
    the download function will update the download status and then revisit if the download is failed or not complete
    '''
    image_front = None
    image_back = None
    download_status = None

    def __init__(self, id, image_front, image_back):
        self.id = id
        self.image_front = image_front
        self.image_back = image_back

    def update_status(self, status):
        self.download_status = status
```

```python
def show_img(data):
    image = np.asarray(bytearray(data), dtype="uint8")

    # use imdecode function
    image = cv2.imdecode(image, 1)
    # img_color = cv2.imread(image,1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
    return image
```

```python
async def fetch(url, file_name, retry=20):
    print(file_name)

    if isinstance(url, str):
        connector = aiohttp.TCPConnector(ssl=False)
        timeout = aiohttp.ClientTimeout(total=30)
        async with aiohttp.ClientSession(connector=connector, timeout=timeout) as session:
            try:
                async with session.get(url) as response:
                    if response.status == 200:
                        data = await response.content.read(-1)
                        # img  = show_img(data)
                        async with aiofiles.open(file_name, "wb") as f:
                            await f.write(data)
                        # return data
            except:
                if retry == 0:
                    return
                print("current try: ", retry)
                if url not in error:
                    print(url)
                    error.append((file_name, url))
                await fetch(url, file_name, retry-1)
    else:
        pass
```

```python
async def download(data, output_folder):
    thread_times = []
    tasks = []
    for item in data:
        t1 = timer()
        task_get_front = asyncio.create_task(fetch(item.image_front, f"{output_folder}/{item.id}_front.jpg"))
        task_get_back = asyncio.create_task(fetch(item.image_back, f"{output_folder}/{item.id}_back.jpg"))
        print("current Item ID ", item.id)
        # if front_result:
        #     img = show_img(front_result)
        #     cv2.imwrite(f"{item.id}_front.jpg", img)

        # if back_result:
        #     img  = show_img(back_result)
        #     cv2.imwrite(f"{item.id}_back.jpg", img)
        t2 = timer()
        thread_times.append(t2 - t1)
        tasks.append(task_get_front)
        tasks.append(task_get_back)
    results = await asyncio.gather(*tasks)

    return thread_times
```
